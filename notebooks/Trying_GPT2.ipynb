```
import os
sess = gpt2.start_tf_sess(threads=4)
# prev check point
# checkpoint_model-3.0.tar
check_dir = os.path.dirname('/content/drive')

gpt2.finetune(sess,
              dataset=file_name,
              model_name='355M',
              restore_from='latest',
              checkpoint_dir=check_dir,
              steps=3000,
              multi_gpu=True,
              learning_rate=0.002,
              run_name='model-3.0',
              print_every=10,
              sample_every=200,
              save_every=500
              )
```
